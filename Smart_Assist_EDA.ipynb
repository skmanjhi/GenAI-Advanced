```python
# prompt: generate a code to read csv files from folder rsics_dataset, concatenate or union  all the csv into one data frame df_1 as every csv has same columns

import pandas as pd
import os
folder_path = "/content/rsics_dataset"
def concatenate_csv_files(folder_path):
    """
    Reads all CSV files from a folder, concatenates them into a single DataFrame,
    and returns the DataFrame.
    """
    all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    df_1 = pd.DataFrame()

    for file in all_files:
      file_path = os.path.join(folder_path, file)
      try:
          temp_df = pd.read_csv(file_path)
          df_1 = pd.concat([df_1, temp_df], ignore_index=True)
      except pd.errors.EmptyDataError:
          print(f"Warning: Skipping empty file {file}")
      except pd.errors.ParserError:
          print(f"Warning: Skipping file with parsing error {file}")
      except Exception as e:
          print(f"Error reading {file}: {e}")

    return df_1

# Example usage (replace 'rsics_dataset' with your actual folder name)
df_1 = concatenate_csv_files(folder_path)
print(df_1.shape)
df_1.head() # Display first 5 rows of the resulting dataframe
```

```python
import torch
from transformers import pipeline

model_id = "meta-llama/Llama-3.2-1B-Instruct"
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]
outputs = pipe(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```
